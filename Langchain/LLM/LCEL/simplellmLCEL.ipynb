```python
### Open AI API Key and Open Source Model LLMs --> Llama3,Gemma2
import os
from dotenv import load_dotenv
load_dotenv()

import openai
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
model=ChatOpenAI(model_name="gpt-4")
model
```

```python
!pip install langchain_core
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
messages = [
    SystemMessage(content="You are a helpful assistant that translates Indonesian to French."),
    HumanMessage(content="Translate the following Indonesian text to French: 'Halo, bagaimana kabarmu?'")
]

result = model.invoke(messages)
```

```python
result
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
## using lcel-chain the components 
chain=model|parser
chain.invoke(messages)
```

```python
from langchain_core.prompts import ChatPromptTemplate

generic_template = "Translate the following into {language}:"

prompt = ChatPromptTemplate.from_messages([
    ("system", generic_template),
    ("user", "{text}")
])
```

```python
result = prompt.invoke({
    "language":"Indonesian",
    "text":"Hello, how are you?"})
```

```python
result.to_messages()
```

```python
chain=prompt|model|parser
chain.invoke({
    "language":"Indonesian",
    "text":"Hello, how are you?"})
```
