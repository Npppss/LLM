```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#Sentence to be encoded
message = ["RNNs are a type of neural network that is well-suited to processing sequences of data."]
```

```python
message
```

```python
#Defining vocabulary size
vocab_size = 10000
```

```python
# One-hot encoding representation
onehot_repr = [one_hot(words, vocab_size) for words in message]
onehot_repr
```

```python
# word emmbedding repsentation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 7
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary() 
```

```python
model.predict(embedded_docs[0].reshape(1,sent_length))
```
