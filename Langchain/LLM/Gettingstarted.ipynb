```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
## Langsmith Tracking
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGSMITH_ENDPOINT"]=os.getenv("LANGSMITH_ENDPOINT")
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-4o-mini")
print(llm)
```

```python
## Input and get response form LLM

result=llm.invoke("Apa itu generative AI?")
```

```python
print(result)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helpful assistant." ),
        ("user","{input}" )
    ]
)

prompt
```

```python
chain=prompt|llm

responses=chain.invoke({"input":"Jelaskan tentang AI dalam 2 paragraf dengan bahasa yang mudah dimengerti oleh anak SD."})
print(responses)
```

```python
type(responses)
```

```python
from langchain_core.output_parsers import StrOutputParser
output_response=StrOutputParser()
chain=prompt|llm|output_response

responses=chain.invoke({"input":"Jelaskan tentang AI dalam 2 paragraf dengan bahasa yang mudah dimengerti oleh anak SD."})
print(responses)
```
